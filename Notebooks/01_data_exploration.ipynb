### Observation 1

The dataset contains 7,043 customer records and 21 features. Each row represents an individual telecom customer, while the columns capture customer demographics, subscription details, account information, and the target variable (churn).

### Observation 2

The dataset contains 3 numerical variables and 17 categorical variables. While most categorical variables are correctly represented as strings, two issues were identified:

1.  "customerID" is an identifier and does not carry predictive value, hence should be excluded from modeling.
2.  "TotalCharges" is incorrectly stored as a string instead of a numeric type, likely due to missing or malformed entries. This will require data cleaning and conversion before further analysis.

Overall, the dataset is well-structured for churn prediction but requires preprocessing to correct data types and handle inconsistencies.

### Observation 3

The dataset does not contain any explicitly recorded missing values, as all columns show 0% null entries. However, this result should be interpreted with caution, as the TotalCharges column is stored as a string type, suggesting potential hidden missing values represented as non-numeric or blank entries. This indicates that further data cleaning is required to ensure accurate numerical analysis.

### Observation 4
A data quality issue was discovered in the TotalCharges column, where 11 records contain blank values instead of numeric entries. Further investigation revealed that all affected records correspond to customers with zero tenure, indicating they are newly registered customers who have not yet incurred charge

### Observation 5

A duplicate check was performed to ensure data integrity. The dataset contains no duplicate records, indicating that each row represents a unique customer.

### Observation 6


Summary statistics reveal a well-structured dataset with meaningful variation in customer behavior. Tenure ranges from 0 to 72 months, indicating both new and long-term customers, while monthly charges vary significantly, suggesting multiple pricing tiers.

Categorical analysis shows a strong dominance of month-to-month contracts and fiber optic internet service, both of which may be linked to churn behavior. The target variable is imbalanced, with a higher proportion of non-churned customers.

A key data quality concern was identified in the TotalCharges column, which shows inconsistent unique values and previously confirmed type issues, requiring further cleaning before modeling.
