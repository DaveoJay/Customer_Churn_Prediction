# Customer Churn Prediction

## 02 - Data Cleaning & Preparation

This notebook prepares the dataset for exploratory analysis and machine learning.

Cleaning activities include:

- Correcting data types
- Handling missing values
- Creating business-relevant indicators
- Removing unnecessary variables

```python
import pandas as pd

from pathlib import Path

BASE_DIR = Path().resolve().parent 

file_path = BASE_DIR / "data" / "raw" / "Telco-Customer-Churn.csv"

df = pd.read_csv(file_path)

df.head()
```

## Convert TotalCharges to Numeric

The `TotalCharges` column is currently stored as an object because it contains blank string values.

Converting this field to numeric format allows it to be used for analysis and machine learning.

```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df.isnull().sum()
```

### Observation

Blank values have now been converted into proper missing values (`NaN`), making them easier to handle consistently.

## Handle Missing Values

Customers with missing `TotalCharges` all have zero tenure and therefore have not yet accumulated charges.

Replacing these values with 0 accurately reflects their business meaning.

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df["TotalCharges"].isnull().sum()
```

## Create New Customer Indicator

Customers with zero tenure represent a unique customer segment and may behave differently from long-term customers.

A binary indicator is created to explicitly capture this information.

```python
df["IsNewCustomer"] = (df["tenure"] == 0).astype(int)

df["IsNewCustomer"].value_counts()
```

## Remove Non-Predictive Features

The `customerID` column serves only as a unique identifier and does not contain information useful for prediction.

Removing it simplifies the dataset and prevents unnecessary noise during modeling.

```python
df.drop("customerID", axis=1, inplace=True)
```

```python
df.head()
```

## Final Validation

Before proceeding, we verify that the dataset contains no remaining missing values.

```python
df.isnull().sum().sum()
```

### Validation Result

The dataset is now fully cleaned and contains no missing values.

## Save Cleaned Dataset

The cleaned dataset will be stored for downstream analysis and feature engineering.

```python
output_path = BASE_DIR / "data" / "processed" / "cleaned_churn.csv"

df.to_csv(output_path, index=False)

print("Cleaned dataset saved successfully.")
```

# Data Cleaning Summary

### Cleaning Actions Performed

- Converted `TotalCharges` to numeric format.
- Handled missing `TotalCharges` values.
- Created `IsNewCustomer`.
- Removed `customerID`.
- Verified dataset integrity.

### Result

The dataset is now analysis-ready and can be used for exploratory data analysis and feature engineering.
